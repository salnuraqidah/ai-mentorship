```python
data = pd.read_csv('data_bar.csv')
data.head(n=10)
```

```python
data.info()
```

```python
from collections import Counter
```

```python
c = Counter(['Python', 'C++', 'Java', 'Python', 'JavaScript'])
c
```

```python
c.update(['C++', 'Python'])
c
```

```python
ids = data['Responder_id']
lang_response = data['LanguagesWorkedWith']

lang_counter = Counter()
```

```python
for response in lang_response:
    lang_counter.update(response.split(';'))
print(lang_counter)    
```

```python
print(lang_counter.most_common(15))
```

```python
lang = []
popularity = []

for item in lang_counter.most_common(15):
    lang.append(item[0])
    popularity.append(item[1])    
print(lang)
print(popularity)
```

```python
plt.bar(lang, popularity)

plt.title('Most Popular language')
plt.xlabel('Language')
plt.ylabel('Users')
plt.show()
```

```python
plt.barh(lang, popularity)

plt.title('Most Popular language')
plt.xlabel('Language')
plt.show()
```

```python
popularity.reverse()
lang.reverse()
plt.barh(lang, popularity)

plt.title('Most Popular Language')
plt.xlabel('Language')
plt.show()
```
