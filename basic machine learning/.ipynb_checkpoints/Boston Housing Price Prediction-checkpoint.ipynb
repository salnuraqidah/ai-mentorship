## Linear Regression

```python
# initialiaze the linear regression model
reg = linear_model.LinearRegression()
```

```python
df_x = df_boston.drop('medv', axis=1)
df_y = df_boston['medv']
print(f"Features: {df_x.shape}")
print(f"Target: {df_y.shape}")
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=42)
print(f"Train features: {x_train.shape}")
print(f"Train target: {y_train.shape}")
print(f"Test features: {x_test.shape}")
print(f"Test target: {y_test.shape}")
```

```python
# split the data into 80% training and 20% testing data
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size = 0.2, random_state = 43)
```

```python
# train the model with our testing data
reg.fit(x_train, y_train)
```

```python
# R-squared
accuracy_score = reg.score(X=df_x, y=df_y)
accuracy_score
```

```python
coeff_df = pd.DataFrame(reg.coef_, df_x.columns, columns=['Coefficient'])
coeff_df
```

```python
# intercept
print(reg.intercept_)
```

```python
y_pred_test = reg.predict(x_test)
y_pred_train = reg.predict(x_train)
```
