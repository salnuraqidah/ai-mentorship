```python
data = pd.read_csv("My Uber Drives - 2016.csv")
data
```

### > Exercise 4

1. Find all trips that is greater than 10 miles and originated from New York and Morris

```python
x = data.loc[(data["MILES*"] > 10 ) & (data["START*"].isin(["New York","Morrisville"]))]
x.reset_index(inplace = True, drop = True)
x
```

### Exercise 5

1. Create a new column with the following condition:\
    a) >10    : Long Trip\
    b) 5-10   : Medium Trip\
    c) <5     : Short Trip

```python
data["DISTANCE"] = np.where(data["MILES*"] < 5, 'Short Trip', 
         (np.where(data["MILES*"] < 10, 'Medium Trip', 'Long Trip')))
data.head(n=10)
```
